```python
from pyspark.sql.functions import col, current_timestamp

raw_path = "/Volumes/f1_project/default/f1_all_csv/f1_data/" 

f1_files = {
    "circuits.csv": "circuits",
    "constructors.csv": "constructors",
    "drivers.csv": "drivers",
    "races.csv": "races",
    "results.csv": "results",
    "pit_stops.csv": "pit_stops",
    "lap_times.csv": "lap_times",
    "qualifying.csv": "qualifying",
    "driver_standings.csv": "driver_standings",
    "constructor_standings.csv": "constructor_standings",
    "seasons.csv": "seasons",
    "sprint_results.csv": "sprint_results",
    "status.csv": "status",
    "constructor_results.csv": "constructor_results"
}


for file_name, table_name in f1_files.items():
    print(f"Ingesting {file_name}...")
    
    df = spark.read.option("header", True).option("inferSchema", True).csv(f"{raw_path}{file_name}")
    

    for column in df.columns:
        df = df.withColumnRenamed(column, column.lower().replace("id", "_id"))
        
    df_final = df.withColumn("ingested_at", current_timestamp())
    
    df_final.write.mode("overwrite").format("delta").saveAsTable(f"f1_processed.{table_name}")

print("All 14 files are now ready in the f1_processed schema!")
```
